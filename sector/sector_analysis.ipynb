```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the years to plot
years = [2021, 2022, 2023]

for year in years:
    # Create a new figure for each year
    fig, ax = plt.subplots(figsize=(13, 8))
    
    # Calculate trade balance and export growth for the specific year
    merge_df[f'trade_balance_{year}'] = merge_df[f'exp_{year}'] - merge_df[f'imp_{year}']
    
    if year == 2021:
        merge_df[f'exp_growth_{year}'] = 0  # Set the base year growth to 0
    else:
        merge_df[f'exp_growth_{year}'] = ((merge_df[f'exp_{year}'] - merge_df[f'exp_{year-1}']) / merge_df[f'exp_{year-1}']) * 100
    
    # Plot for the specific year
    scatter = sns.scatterplot(
        ax=ax,
        x=f'trade_balance_{year}', 
        y=f'exp_growth_{year}', 
        size=f'exp_{year}', 
        hue='SELETED COMMODITIES', 
        data=merge_df, 
        sizes=(100, 2000), 
        palette='Set2'
    )

    # Titles and labels
    ax.set_title(f'Sector Analysis: {year}', fontsize=14)
    ax.set_xlabel('Trade Balance (Millions)', fontsize=12)
    ax.set_ylabel('Export Growth (%)', fontsize=12)
    
    # Add a legend to the plot
    ax.legend(loc='upper left', bbox_to_anchor=(1, 1), title="Selected Commodities")

    # Adjust layout
    plt.tight_layout(rect=[0, 0, 0.9, 1])
    
    # Show the plot
    plt.show()
```

```python
sector_summary = merge_df.groupby('SECTOR').sum()

# Plotting
fig, ax = plt.subplots(figsize=(14, 7))

# Trade Balance Plot
sector_summary[['trade_balance_2021', 'trade_balance_2022', 'trade_balance_2023']].plot(ax=ax, marker='o')

# Set title and labels
ax.set_title('Trade Balance by Sector (2021-2023)', fontsize=14)
ax.set_ylabel('Trade Balance (Millions)', fontsize=12)
ax.set_xlabel('Sector', fontsize=12)
ax.grid(True)

# Set x-ticks to sector names
ax.set_xticks(range(len(sector_summary.index)))
ax.set_xticklabels(sector_summary.index)

for i, sector in enumerate(sector_summary.index):
    ax.text(i, sector_summary['trade_balance_2021'][sector], f"{sector_summary['trade_balance_2021'][sector]:,.0f}", 
            ha='center', va='bottom', fontsize=7, fontweight='bold')
    ax.text(i, sector_summary['trade_balance_2022'][sector], f"{sector_summary['trade_balance_2022'][sector]:,.0f}", 
            ha='center', va='bottom', fontsize=7, fontweight='bold')
    ax.text(i, sector_summary['trade_balance_2023'][sector], f"{sector_summary['trade_balance_2023'][sector]:,.0f}", 
            ha='center', va='bottom', fontsize=7, fontweight='bold')

# Supertitle
plt.suptitle('Sector Performance Analysis: Agriculture, Mining, Manufacture (2021-2023)', fontsize=16)

# Adjust layout to avoid overlap
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show the plot
plt.show()
```
